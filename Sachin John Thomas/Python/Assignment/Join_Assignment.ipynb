```python
import pandas as pd
```

```python

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aliyamma', 'Bobby', 'Charlie', 'Dave'],
    'Age': [25, 30, 35, 40]
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'City': ['Adoor', 'Pallimukku', 'Chicago', 'Indore'],
    'Salary': [50000, 60000, 70000, 80000]
})
```

```python
inner_join = pd.merge(df1, df2, on='ID', how='inner')
inner_join
```

```python
# Left join (all IDs from df1, missing values from df2)
left_join = pd.merge(df1, df2, on='ID', how='left')
left_join
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')
right_join
```

```python
outer_join = pd.merge(df1, df2, on='ID', how='outer')
outer_join
```
